## Import the Necessary Libraries

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering

import warnings
warnings.filterwarnings('ignore')
```

## Import the Data

```python
crime_details= pd.read_csv('crime_data.csv')
crime_details.head()
```

```python
crime_details.shape
```

### Hierarchical Clustering

```python
# Normalized the data function
def norm_fun(i):
    x= ((i-i.min())/(i.max()- i.min()))
    return(x)
```

```python
norm_crimedata= norm_fun(crime_details.iloc[: ,1:])
norm_crimedata.head()
```

```python
plt.figure(figsize=(15,6))
dendogram= sch.dendrogram(sch.linkage(norm_crimedata, method='average'))
```

```python
#Create Clustering

agg_clust=AgglomerativeClustering( n_clusters=3,affinity='euclidean',linkage='complete')
```

```python
y_clust= agg_clust.fit_predict(norm_crimedata)
```

```python
crime_details['clusters_Hier']= y_clust
crime_details
```

```python
crime_details['clusters_Hier'].value_counts()
```

### K Means

```python
from sklearn.cluster import KMeans 
from scipy.spatial.distance import cdist
```

```python
crime_details_2= pd.read_csv('crime_data.csv')
crime_details_2.head()
```

```python
# Normalized data function 

def norm_fun (i):
    x= (i-i.min())/(i.max()-i.min())
    return (x)
```

```python
norm_crimedata2= norm_fun(crime_details_2.iloc[:,1:])
norm_crimedata2
```

```python
# Ploting the ELbow Curve

wcss=[]

for i in range (1,12):
    kmeans= KMeans(n_clusters=i)
    kmeans.fit(norm_crimedata2)
    wcss.append(kmeans.inertia_)

plt.plot(range(1,12), wcss, 'o-')
plt.title("ELbow Curve")
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
#selecting 4 CLusters from above ELbow curve
model= KMeans( n_clusters=4)
k_model= model.fit(norm_crimedata2).labels_
```

### DBSCAN

```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
crime_details3= pd.read_csv('crime_data.csv')
crime_details3.head()
```

```python
crime_dbscan= crime_details3.iloc[:,1:]
crime_dbscan.values
```

```python
scaler= StandardScaler().fit_transform(crime_dbscan.values)
scaler
```

```python
model3= DBSCAN(eps=2,min_samples=5,metric='euclidean')
```

```python
descan_cluster= model3.fit(scaler).labels_
```

```python
crime_details3['CLuster']= descan_cluster
```

```python
crime_details3
```

```python
crime_details3['CLuster'].value_counts()
```

```python
# Plot between pairs Murder~Assault
sns.lmplot('Assault','Rape',  data=crime_details3, hue = 'CLuster',fit_reg=False, size = 6 )
```
